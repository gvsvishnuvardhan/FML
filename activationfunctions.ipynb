```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid_fun(x):
  return 1 / (1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid_fun(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
```

```python
plt.show
```

```python
def tanh(x):
  return np.tanh(x)
```

```python
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('Activation Function :tanh')
plt.show
```

```python
def relu(x):
  l=[]
  for i in x:
    if i<0:
      l.append(0)
    else:
      l.append(i)
  return l
plt.plot(x,relu(x))
plt.title('Activation Fun : relu')
plt.axis('tight')
plt.show
```
